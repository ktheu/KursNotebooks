#### Verbesserung

```python
var x;
var xspeed;
var y;
var yspeed;
var lost;
var punkte;
var xrechteck;
var highscore = 0;

function setup() { 
    createCanvas(1080, 720); 
    colorMode(HSB, 360, 100, 100) 
    initValues();
} 

function draw() {

    // ----------------------- check ---------------------------
    var kollisionLinksRechts = (x + 16 > width) || (x - 16 < 0);
    var kollisionOben = y - 16 < 0;
    var abprallen = (y + 16 > 620) && (x +16 >= xrechteck) && (x -16 <= xrechteck + 150);
    lost = y >= height + 32;

    // ------------------------- move -------------------------
    // Bewegung Ball
    xspeed = xspeed * 1.0003
    yspeed = yspeed * 1.0003

    if (kollisionLinksRechts) xspeed = -xspeed;
    if (kollisionOben) yspeed = -yspeed;

    if (abprallen) {
        punkte = punkte + 1;
        yspeed = -yspeed;
    }
    
    x = x + xspeed 
    y = y + yspeed 

    // Bewegung Rechteck
    xrechteck = mouseX
    if (xrechteck >= 930) {
        xrechteck = 930;
    }

    // --------------------- draw ---------------------------------
    // Hintergrund
    fill((y + 300) / 7, 50, 90)  //weiß (0,0,100) blau(180,y/9,70)  gelb(y+300)/7,50,90
    rect(0, 0, 1080, 720);

    // Ball
    fill(y / 7, 100, 100)
    noStroke()
    ellipse(x, y, 32, 32)

    // Rechteck
    noStroke();
    fill(y / 7, 100, 100)
    rect(mouseX, 620, 150, 50)

    // Text
    fill(120, 100, 70);
    textSize(40);
    text("Punkte:", 10, 50);
    text(punkte, 160, 50);

    if (lost) {
        fill(60, 10, 85);
        textSize(90);
        text("Game Over", 300, 360);
        fill(16, 90, 0);
        textSize(40);
        text("Punkte:" + punkte, 440, 440);
        text("Maustaste drücken zum Neustarten", 220, 500)
        if (punkte > highscore) {
            highscore = punkte;
        }
        text("Highscore:" + highscore, 440, 650);
        noLoop();
    }
}

function mousePressed() {
    if (lost) {
        initValues();
        loop();
    }
}

function initValues() {
    x = int(random(20, 1060));
    xspeed = 7;
    y = 20;
    yspeed = 5;
    punkte = 0
    xrechteck = mouseX;
}
```

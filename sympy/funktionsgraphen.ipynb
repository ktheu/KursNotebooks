## Schaubilder von Funktionen

Die *plot* Funktion von sympy zeigt Graphen im Bereich zwischen -10 und 10. Bei Polstellen ist die Ausgabe mit matplotlib besser.

```python
from sympy import *
from sympy.plotting import plot
%matplotlib inline

x = symbols('x', real=True)
p1 = plot(x*x)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(0.9,1.1, 0.0001)
y = -3/(x-1) - 2

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()

plt.ylim((-500,500))
plt.show()
```

### Ganzrationale Funktionen

Skizziere die Schaubilder folgender Funktionen und bestimme die Schnittpunkte mit den Koordinatenachsen.

a. $f(x) = \frac{1}{2}x +1$ <br>
b. $f(x) = (x-1)^2 -4$ <br>
c. $f(x) = (x-1)^3 + 1$

```python
# a
expr = Rational(1,2)*x + 1
p = plot(expr)
print('Schnitt mit x-Achse: {} , Schnitt mit y-Achse {}'.format(solve(expr) , expr.subs(x,0)))
```

```python
# b
expr = (x-1)**2 - 4
p = plot(expr, xlim=(-3,4), ylim=((-5,5)))
print('Schnitt mit x-Achse: {} , Schnitt mit y-Achse {}'.format(solve(expr) , expr.subs(x,0)))
```

```python
#c 
expr = (x-1)**3 + 1
p = plot(expr, xlim=(-3,4), ylim=((-5,5)))
print('Schnitt mit x-Achse: {} , Schnitt mit y-Achse {}'.format(solve(expr) , expr.subs(x,0)))
```

### Potenzfunktionen

Skizziere die Schaubilder folgender Funktionen und bestimme jeweils die Asymptoten.

a. $f(x) = \frac{1}{x+1}+2$  <br>
b. $f/x) = -\frac{3}{x-1} -2$ <br>
c. $f(x) = -\frac{2}{(x+1)^2}$

```python
# a
 
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(-1.1,-0.9, 0.0001)  # ab hier ist x kein sympy symbol
y = 1/(x+1) + 2

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()

plt.ylim((-200,200))
plt.show()

print('Horizontale Asymptote: {}, Vertikale Asymptote: {}'.format("y=0","x=-1"))
```

```python
# b
%matplotlib inline

x = np.arange(0.5,1.5, 0.0001)  # ab hier ist x kein sympy symbol
y = -3/(x-1) - 2

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()

plt.ylim((-200,200))
plt.show()

print('Horizontale Asymptote: {}, Vertikale Asymptote: {}'.format("y=0","x=1"))
```

```python
# c

x = np.arange(-3, 1, 0.001)  # ab hier ist x kein sympy symbol
y = -2/(x+1)**2

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()

plt.ylim((-300,50))
plt.show()

print('Horizontale Asymptote: {}, Vertikale Asymptote: {}'.format("y=0","x=-1"))
```

### Trigonometrische Funktionen

Skizzieren die Schaubilder folgender Funktionen und gib jeweils die Periode an.

a. $f(x) = \sin(x)$ <br>
b. $f(x) = -sin(2x) + 1$ <br>
c. $f(x) = sin(\frac{1}{2}\pi(x+1))$

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 500)
y = np.sin(x)
fig, ax = plt.subplots(figsize=(12, 3))
ax.plot(x, y, lw=2)
ax.set_yticks([-1, 0, 1])
ax.set_xticks([-2 * np.pi, -np.pi, 0, np.pi, 2 * np.pi])
ax.set_xticklabels([r'$-2\pi$', r'$-\pi$', 0, r'$\pi$',r'$2\pi$'])
x_minor_ticker = mpl.ticker.FixedLocator([-3 * np.pi / 2, -np.pi / 2, 0, np.pi / 2, 3 * np.pi / 2])
ax.xaxis.set_minor_locator(x_minor_ticker)
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

print('Periode: {}'.format('2 pi'))
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 500)
y = -np.sin(2*x)+1
fig, ax = plt.subplots(figsize=(12, 3))
ax.plot(x, y, lw=2)
ax.set_yticks([-1, 0, 1, 2, 3])
ax.set_xticks([-2 * np.pi, -np.pi, 0, np.pi, 2 * np.pi])
ax.set_xticklabels([r'$-2\pi$', r'$-\pi$', 0, r'$\pi$',r'$2\pi$'])
x_minor_ticker = mpl.ticker.FixedLocator([-3 * np.pi / 2, -np.pi / 2, 0, np.pi / 2, 3 * np.pi / 2])
ax.xaxis.set_minor_locator(x_minor_ticker)
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

print('Periode: {}'.format('pi'))
```

```python
x = np.linspace(-6, 6, 500)
y = np.sin(1/2*np.pi*(x+1)) 
fig, ax = plt.subplots(figsize=(12, 3))
ax.plot(x, y, lw=2)
ax.set_yticks([-1, 0, 1])
ax.set_xticks(range(-6,7,1))
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

print('Periode: {}'.format('4'))
```

### Exponentialfunktionen

Skizziere das Schaubild folgender Funktionen und bestimme jeweils die Asymptote.

a. $f(x) = \mathrm{e}^{x-1} + 1$ <br>
b. $f(x) = \mathrm{e}^{-(x-1)} + 2$ <br>
c. $f(x) = \mathrm{e}^{-x+3}+1$

```python
x = np.linspace(-4, 3, 500)
y = np.e ** (x - 1) + 1 
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(x, y, lw=2)
#ax.set_yticks([-1, 0, 1])
ax.set_xticks(range(-4,4,1))
ax.set_yticks(range(0,10,1))
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

print('Asymptote: {}'.format('y=1'))
```

```python
x = np.linspace(-1, 5, 500)
y = np.e ** -(x - 1) + 2
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(x, y, lw=2)
#ax.set_yticks([-1, 0, 1])
ax.set_xticks(range(-1,5,1))
ax.set_yticks(range(0,10,1))
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

print('Asymptote: {}'.format('y=2'))
```

```python
x = np.linspace(1, 8, 500)
y = np.e ** (-x + 3) + 1 
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(x, y, lw=2)
#ax.set_yticks([-1, 0, 1])
ax.set_xticks(range(-1,9,1))
ax.set_yticks(range(0,10,1))
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

print('Asymptote: {}'.format('y=1'))
```

### Darstellung einer Funktion

```python
import numpy as np
import matplotlib.pyplot as plt

# evenly sampled time at 200ms intervals
x = np.arange(-5., 5., 0.2)

plt.plot(x, x**3, 'red')
plt.xlabel("x")
plt.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
plt.show()
```

```python
# Loss-Funktion der logistischen Regression
def L(y,a):
    return -(y*np.log(a)+(1-y)*np.log(1-a))

a = np.arange(0.1, 1., 0.01)
plt.plot(a, L(1,a), 'red')
plt.plot(a, L(0,a), 'blue')
plt.plot(a, L(0.5,a), 'yellow')
plt.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
plt.show()
```

```python
anzahl = 50
w1, w2, b = 1, -1, 0.5

X1 = [5*random.random() for _ in range(anzahl)] 
X2 = [5*random.random() for _ in range(anzahl)] 
C = [w1*X1[i]+ w2*X2[i] +b > 0 for i in range(anzahl)]

plt.scatter(X1, X2, c=C, s=30)

plt.axis([0, 5, 0, 5])
plt.show()
```
